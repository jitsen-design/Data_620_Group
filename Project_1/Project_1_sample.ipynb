```python
import json
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("beer.csv")
df.head()
```

```python
df.shape
```

```python
for col in ['beer_name','review_profilename']:
    print(col, ', unique:', df[col].nunique())
```

```python
G = nx.DiGraph()
for index, row in df.iterrows():
    G.add_node(row['review_profilename'], label='reviewer')
    G.add_node(row['beer_name'], beer_style=row['beer_style'], label='beer')
    G.add_edge(row['review_profilename'], row['beer_name'], weight=row['review_overall'])
```

```python
for style in df['beer_style'].unique():
    
    subgraph = nx.DiGraph()
    
    selected_nodes = [(a,b) for a,b in G.edges() 
                   if G.nodes(data=True)[b]['beer_style'] == style]
    
    subgraph.add_edges_from(selected_nodes)
    
    centrality = nx.eigenvector_centrality(subgraph,
                                           max_iter=1000)
    
    print('Style: {}'.format(style))
    print('---------------------------------------------------------------')
    print(sorted(centrality.items(), key=lambda x: x[1], reverse=True)[:3])
```
